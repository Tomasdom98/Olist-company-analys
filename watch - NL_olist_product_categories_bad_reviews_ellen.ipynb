````python
# Build a dataset merging the 3 datasets

df = product_category.merge(orders_data)
df
````

````python
df=df.merge(reviews_data)
df
````

## Data cleaning

````python
# Remove reviews for orders delivered BEFORE expected

df=df[df['order_delivered_customer_date']<df['order_estimated_delivery_date']]


# Remove reviews about undelivered orders 
df[df['order_status']== 'delivered']
df
````

````python
# Keep only text columns and review score

df=df[['order_id','review_score','product_category_name','review_comment_title','review_comment_message' ]]
df
````

````python
df = df.dropna(subset=['review_comment_title','review_comment_message'])
````

````python
df.info()
````

````python
df
````

````python
# combine review title and review message
df['comment_title']=df['review_comment_title']+ ' ' + df['review_comment_message']
df.info()
````

````python
# Clean reviews text
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords 
import string
import unidecode


def clean (text):
    
    for punctuation in string.punctuation:
        text = text.replace(punctuation, ' ') # Remove Punctuation
        
    lowercased = text.lower() # Lower Case
    
    unaccented_string = unidecode.unidecode(lowercased) # remove accents
    
    tokenized = word_tokenize(unaccented_string) # Tokenize
    
    words_only = [word for word in tokenized if word.isalpha()] # Remove numbers
    
    stop_words = set(stopwords.words('portuguese')) # Make stopword list
    
    without_stopwords = [word for word in words_only if not word in stop_words] # Remove Stop Words
    
    return " ".join(without_stopwords)

df['clean_text'] = df['comment_title'].apply(clean)

df.head()
````

## Bad reviews analysis per category

````python
# Groupby product category and aggregate mean, min, max review scores
bad_review=df.groupby(by='product_category_name').agg({'review_score':['count','mean', 'min', 'max']}).sort_values([('review_score','mean')],ascending=False)
bad_review
````

````python
# Keep categories that have more than 100 reviews
bad_review=bad_review[bad_review[('review_score','count')]>=100]
bad_review
````

````python
# Filter out the relogios_presentes category

df
````

````python
bad_review_watch=df[df['product_category_name']=='relogios_presentes']
bad_review_watch
````

````python
# Filter out bad reviews
bad_review_watch=bad_review_watch[bad_review_watch['review_score']==1]
bad_review_watch
````

````python
# Encode bad reviews text
#128 samples 
from sklearn.feature_extraction.text import TfidfVectorizer

# Tuned TFidfvectorizer
tf_vec = TfidfVectorizer(ngram_range = (2,2), min_df=0.01, max_df = 0.05)
vec= tf_vec.fit(bad_review_watch.clean_text)
# Transform text to vectors
vectors=vec.transform(bad_review_watch.clean_text)
````

````python
vectors.toarray().shape #
````

````python
vec.vocabulary_
````

````python
# Sum of tfidf weighting by word
sum_tfidf=vectors.sum(axis=0)
sum_tfidf=sum_tfidf.tolist()[0]
sum_tfidf
````

````python
vec.vocabulary_.keys()
````

````python
# Get the word and associated weight

val=list(zip(vec.vocabulary_.keys(), sum_tfidf ))

import pandas as pd
keywords_df=pd.DataFrame(val, columns=['keywords','tfidf'])
keywords_df
````

````python
# Sort
keywords_df=keywords_df.sort_values('tfidf', ascending=False, ignore_index=True)
keywords_df
````

````python
# Display the sorted list
fake_list=keywords_df[:10]['keywords']
fake_list
````
